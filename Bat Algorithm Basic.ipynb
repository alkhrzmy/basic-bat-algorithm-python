```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective_function(x):
    return x[0]**2 + x[1]**2
```

```python
def initialize_bats(num_bats, num_dimensions):
    bats = []
    for _ in range(num_bats):
        position = np.random.uniform(-10, 10, num_dimensions)
        frequency = np.random.uniform(0, 1)
        bats.append((position, frequency))
    return bats
```

```python
def bat_algorithm_optimization(num_bats, num_dimensions, max_iterations):
    bats = initialize_bats(num_bats, num_dimensions)
    best_position = bats[0][0]
    best_fitness = float('inf')
    
    for _ in range(max_iterations):
        for i, (position, frequency) in enumerate(bats):
            # Generate a random solution
            random_solution = np.random.uniform(-10, 10, num_dimensions)
            
            # Update bat position and velocity
            velocity = np.add(position, np.multiply((position - best_position), frequency))
            position = np.add(position, velocity)
            
            # Update frequency and loudness
            frequency *= 0.9
            loudness = 1 - np.exp(-0.1 * _)
            
            # Evaluate fitness of the new position
            fitness = objective_function(position)
            
            # Update the best position and fitness
            if fitness < best_fitness:
                best_position = position
                best_fitness = fitness
            
            # Update the position of the bat
            bats[i] = (position, frequency)
    
    return best_position, best_fitness
```

```python
def plot_convergence_curve(convergence_curve):
    plt.plot(range(len(convergence_curve)), convergence_curve)
    plt.xlabel('Iteration')
    plt.ylabel('Best Fitness')
    
    plt.title('Convergence Curve')
    plt.show()
```

```python
num_bats = 10
num_dimensions = 2
max_iterations = 100
```

```python
best_position, best_fitness = bat_algorithm_optimization(num_bats, num_dimensions, max_iterations)
print("Best Position:", best_position)
print("Best Fitness:", best_fitness)
```
